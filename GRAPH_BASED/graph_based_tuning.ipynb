```python
from GRAPH_BASED.graph_based_recommender_alpha import graph_based_recommender_alpha
from utils.evaluator.evaluator import evaluator
from utils.submission_helper import submission_helper
from utils.data_manager.data_manager import data_manager
from utils.data_manager.splitter import splitter
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
training, validation, test = splitter().get_train_evaluation_test(percentage_of_training_data=0.6,
                                                                  percentage_of_validation_data=0.2)
```

```python
alphas = [i for i in np.arange(0.0, 1.0 , 0.1)]
k_vector2 = [20,40]
k_vector = [k for k in range(50,500, 100)]
k_vector = k_vector2 + k_vector
print(len(alphas)*len(k_vector))
map_vector = []
for alpha in alphas:
    for k in k_vector:
        recommender = graph_based_recommender_alpha()
        recommender.fit(training_set=training.copy(),
                        alpha=alpha,
                        k=k)
        map_ = evaluator.evaluate_already_trained(recommender=recommender)[2]
        print(map_)
        map_vector.append(map_)
print(map_vector)
```

```python
couples = []
for alpha in alphas:
    for k in k_vector:
        couples.append([alpha, k])
chosen = couples[np.argmax(map_vector)]
chosen
```
